```python
from sklearn.svm import SVC
from sklearn.datasets import load_breast_cancer 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
X, y = load_breast_cancer(return_X_y=True)
```

```python
X.shape
```

```python
np.unique(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=200)
```

```python
kernels = ['linear', 'poly', 'sigmoid', 'rbf']

for k in kernels:
	clf = SVC(kernel=k, degree=1, gamma='auto').fit(X_train, y_train)
	print("kernel: {}, score: {}".format(k, clf.score(X_test, y_test)))
        
```

```python
pd.DataFrame(X).describe([0.01, 0.1, 0.2, 0.5, 0.7, 0.9, 0.99]).T

#量纲不统一
#部分数据偏态
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
X = StandardScaler().fit_transform(X)
pd.DataFrame(X).describe([0.01, 0.1, 0.2, 0.5, 0.7, 0.9, 0.99]).T
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=200)

kernels = ['linear', 'poly', 'sigmoid', 'rbf']

for k in kernels:
	clf = SVC(kernel=k, degree=1, gamma='auto').fit(X_train, y_train)
	print("kernel: {}, score: {}".format(k, clf.score(X_test, y_test)))
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.model_selection import GridSearchCV
```

```python
cv = StratifiedShuffleSplit(n_splits=5, test_size=0.3, random_state=200)
```

```python
clf = SVC(kernel='poly', degree=1, cache_size=3000)
```

```python
#交叉验证需要的参数 
gamma_range = np.logspace(-9, 1, 20)
coef_range = np.linspace(0, 5, 20)
para_grid = dict(gamma=gamma_range, coef0=coef_range)
para_grid
```

```python
gridSear = GridSearchCV(clf, param_grid=para_grid, cv=cv)
gridSear.fit(X, y)
```

```python
print("best score: {}, respect best param: {}".format(gridSear.best_score_, gridSear.best_params_))
```
